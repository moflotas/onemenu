```python
import pandas as pd

orders = pd.read_csv('restaurant-1-orders.csv')


orders
```

```python
orders['Order Date'] = pd.to_datetime(orders['Order Date'], format="%d/%m/%Y %H:%M")
list = []
months = []
# Filter the DataFrame for July 2019 orders
for x in range (2016, 2020):
    july_2019_orders = orders[(orders['Order Date'].dt.year == x)]
    order_sums = july_2019_orders.groupby('Order Number')['Product Price'].sum()
    list.append(order_sums.mean())
    months.append(x)

list
```

```python
import matplotlib.pyplot as plt
plt.plot(months, list)

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Avg order sum')
plt.title('Change of average order sum through the years')

# Displaying the plot
plt.show()
```

```python
top_dishes = orders['Item Name'].value_counts().nlargest(10)

```

```python
plt.figure(figsize=(10, 6))
top_dishes.plot(kind='bar')
plt.title('Top 10 Dishes')
plt.xlabel('Dish')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
